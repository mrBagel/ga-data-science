```python
# Write a function get_max that takes a list of
# numbers and returns the max of those numbers,
# don't use the builtin max() function. Afterward, try using max()

def get_max(numbers):
    current_biggest = numbers[0]
    for num in numbers:
        if num > current_biggest:
            current_biggest = num
    return current_biggest


print get_max([1, 2, 3, 4, 5])
print get_max([-1, -2, -3, -4, -5, 0])            
```

```python
# We're going to learn how to use sorted() with different 
# score functions
movies = [
    {
        "name": "Shawshank Redemption",
        "metacritic": 9.1,
        "rotten_tomatoes": 8.7,
        "fb_fan_count": 800000
    },
    {
        "name": "The Usual Suspects",
        "metacritic": 8.8,
        "rotten_tomatoes": 8.4,
        "fb_fan_count": 75000
    },
    {
        "name": "Never Say Never",
        "metacritic": 4.3,
        "rotten_tomatoes": 3.3,
        "fb_fan_count": 100000000,
    }
]

best_by_fan_count = sorted(
    movies,
    # key is always a function
    key=lambda movie: movie["fb_fan_count"],
    # descending order; this is False by default
    reverse=True)

print best_by_fan_count
```

```python
# Sort movies by metacritic score, ascending
best_by_metacritic = sorted(
    movies,
    key=lambda movie: movie["metacritic"]
)
print best_by_metacritic
```

```python
# if we wanted to replicate the functionality of counter ourselves
def aggregate_counts(order_list):
    orders_by_count = {}
    for order in order_list:
        if order in orders_by_count:
           orders_by_count[order] += 1
        else:
           orders_by_count[order] = 1
    return orders_by_count
```

```python
aggregate_counts(["apple", "orange", "apple", "bob", "tangerine"])
```

```python
# Here's a function that will clean up our year data and return as integers
def clean_year(year):
    # first, try to cast as integer
    try:
        sanitized_year = int(year)
    except ValueError:
        pass
    else:
        # let's do a sanity check and exclude everything before 1900
        if sanitized_year > 1900:
            return sanitized_year
```
