## Summerize the result
* On an hourly base, there were more people ride during rush hours (8am - 10am, 5pm - 7pm)
* Throughout 2 years in this dataset, the daily number of riders was increasing. Also, there seems to be a strong seasonality pattern that more poeple tend to rider in spring and summer time and less people ride in winter and fall time.
* From all the features given, I selected 'hr', 'season', 'workingday' and 'weathersit' to look at beacuse these features seems to be more representative than other simliar features. From the scatter metrix plot we can see that
    1. more registered users ride during rush hour, while more casual users ride during day time (10am to 5pm).
    2. 'season' doesn't tell much of a difference between registered and causual.
    3. 'wordkingday' is a strong indicator that more registered users ride on working days while more casual users ride on non-working days.
    4. 'weathersit' tells that more registered users tend to ride in bad weathers. Both group tend to ride more on good weather and less in extreme weather.
