## NYC Restaurant Inspection Data Analysis
* Get the data from NYC Open data: https://data.cityofnewyork.us/Health/DOHMH-New-York-City-Restaurant-Inspection-Results/xx67-kt59
* The dataset is inspection results in NYC restaurants. Details include restaurant locations, cusine types, viloations and grades.

##Data Loading and Cleaning up
* I decided to choose most recent year data for my analysis since it will be more relavite to restaurants' most recent hygiene conditions. Also, not every inspection can lead to a grade so there will be many NA values for grades. So my strategy is to simply drops thoese records.
* Some of the cuisine catgories weren't very well organized. For example, "Cafe" and "Coffee/Tea" were treated as different categories, so I consolidated some of the similar categories into one.

```python
df = pd.read_csv('ny_restaurant_inspection.csv')
df.columns = [c.lower() for c in df.columns]
df = df.rename(columns={'cuisine description': 'cuisine', 'inspection date': 'date'})
df['date'] = pd.to_datetime(df['date'])
df.head()
df.shape
```

```python
# Get recent data from 2013 to present and drop missing values
df_recent = df[df['date'] >= '2013-01-01']
df_recent = df_recent.dropna()
df_recent = df[df.boro != 'Missing']
print df_recent.shape
df_recent.head(5)
```

```python
#what cuisines are in the datset
print df_recent.cuisine.unique()
```

```python
#Some of the categories are repetitive such as cafe and cafe/tea, so need to consolidate them
def re_category(t):
    t=t['cuisine']
    if t in ['Caf\xc3\xa9/Coffee/Tea','Caf\xc3\x83\xc2\xa9/Coffee/Tea']:
        return 'Coffee&Tea'
    elif t in ['Sandwiches/Salads/Mixed Buffet','Salads','Soups & Sandwiches','Soups','Soups & Sandwiches']:
        return 'Salad, Sandwich and Soup'
    elif t in ['Hotdogs','Hotdogs/Pretzels','Bagels/Pretzels']:
        return 'Hotdog, Bagel and Pretzel'
    elif t == 'Bottled beverages, including water, sodas, juices, etc.':
        return 'Beverage Shop'
    elif t in ['Chinese', 'Chinese/Cuban', 'Chinese/Japanese']:
        return 'Chinese'
    elif t in ['Mexican', 'Tex-Mex']:
        return 'Mexican'
    elif t in ['Pizza','Italian','Pizza/Italian']:
        return 'Pizza/Italian'
    elif t == 'Latin (Cuban, Dominican, Puerto Rican, South & Central American)':
        return 'Latin'
    elif t in ['Creole/Cajun','Creole','Cajun']:
        return 'Creole/Cajun'
    elif t in ['Juice, Smoothies, Fruit Salads','Fruits/Vegetables']:
        return 'Juice, Smoothies, Fruit, Vege'
    else:
        return t

df_recent['new_category']=df_recent.apply(re_category,axis=1)
```

```python
## Here are the categories for the study
print df_recent.new_category.unique()
df_recent = df_recent[df_recent.new_category != 'Other']
df_recent = df_recent[df_recent.new_category != 'Not Listed/Not Applicable']
```

```python
def new_type(x):
    if x['new_category'] in ['Bakery','Hamburgers','Delicatessen','Ice Cream, Gelato, Yogurt, Ices','Chicken',
             'Donuts','Salad, Sandwich and Soup','Hotdog, Bagel and Pretzel','Steak','Coffee&Tea',
             'Pancakes/Waffles','Soul Food' ,'Seafood','Beverage Shop','Vegetarian','Barbecue','Sandwiches',
             'Juice, Smoothies, Fruit, Vege','Nuts/Confectionary','Tapas']:
        return 'Food Type'
    else:
        return 'Food Origin'

df_recent['new_type']=df_recent.apply(new_type,axis=1)
```

## Exploratory Analysis
* The DOHMH grading systems are based on the followings:
    * A score of less than 14 points on either initial or re-inspection results in an “A” grade
    * On re-inspection, a score of 14-27 points means a restaurant receives both a “B” grade and a “Grade Pending” card
    * On re-inspection, a score of 28 or more points means a restaurant receives both a “C” grade and a “Grade Pending” card
    * When critical violations have not been corrected on re-inspection, a restaurant may receive an additional compliance inspection within one to two weeks, but these inspections do not contribute to a grade.
* I am interested in coverting grade letters to a more quanitive measure since there are will be difference between two restaurants recieved the same grade. So I will convert the grade to a 100 scale and take the average for two years.

```python
#Convert the score to a 100 scale 
df_recent['new_score'] = df_recent.apply(lambda x: 100-x['score'],axis=1)
df_group = df_recent.groupby(['camis','dba','boro','building','street','zipcode','new_category','new_type'])
```

```python
df_new = df_group.new_score.mean().reset_index()
df_new.head()
```

```python
boro = df_new.groupby('boro').new_score.mean()
food_type = df_new[df_new.new_type=='Food Type'].groupby('new_category').new_score.mean()
food_origin = df_new[df_new.new_type=='Food Origin'].groupby('new_category').new_score.mean()
```
